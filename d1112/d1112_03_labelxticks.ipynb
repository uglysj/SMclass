```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Win용
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
## x, y축의 데이터 / 리스트, 튜플, DataFrame - 리스트형태, Series
x = [1, 2, 3]
y = [2, 4, 8]

## 선 그래프 생성
plt.plot(x, y)

## 상단 타이틀
plt.title('라인그래프 Line', size=30) # size: 글자크기 변경

## 그래프 축 설명
# 그래프 X축설명, color: 글자색변경/색상값, 16코드, loc: 글자위치
plt.xlabel('X축 날짜', color='#b3e8b3', loc='center') # left, right, center
# 그래프 Y축설명
plt.ylabel('Y축 인원', color='blue', loc='center') # top, center, bottom

## 그래프 눈금단위 표시
plt.xticks([1, 2, 3])
plt.yticks([2, 3, 4, 8, 9, 10])
```

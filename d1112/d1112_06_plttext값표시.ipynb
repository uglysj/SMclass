```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Win용
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

그래프에 값 표시 plt.text

```python
x = [1, 2, 3]
y = [2, 4, 8]

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='범례데이터', marker='o')
## text(x좌표위치, y좌표위치, 데이터값), ha: 글자가운데정렬
## for
# plt.text(x[0], y[0]+0.2, y[0], ha='center')
for idx, txt in enumerate(y):
  plt.text(x[idx]+0.05, y[idx]-0.25, y[idx], ha='center', color='orange')

plt.title('라인그래프')
plt.legend()
plt.show()
```

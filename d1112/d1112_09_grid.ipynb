```python
import pandas as pd
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Win용
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df[['키', '이름']]
```

그래프 grid: 안내선 출력

```python
plt.plot(df['이름'], df['키'])
plt.xticks(rotation=45)
## 그래프 안내선, axis: x, y 축으로 선이 그려짐, linestyle: 선모양
# linewidth: 선 두께, color: 선색상
plt.grid(axis='y', color='purple', linestyle='--', linewidth=2, alpha=0.2)
plt.grid(axis='x', alpha=0.2)

# 그래프 값 추가
for i in range(8):
  plt.text(df.iloc[i]['이름'], df.iloc[i]['키'], df.iloc[i]['키'], ha='center')
plt.show()
```

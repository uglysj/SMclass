```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Win용
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8] # SM 매출액
y2 = [5, 1, 3] # LG 매출액
y3 = [1, 2, 5] # SK 매출액

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='SM매출액', marker='o')
plt.plot(x, y2, label='LG매출액', marker='^')
plt.plot(x, y3, label='SK매출액', marker='s')

### 그래프 내 값 출력
for i in range(3):
  plt.text(x[i], y[i]+0.2, y[i], ha='center')
  plt.text(x[i], y2[i]+0.2, y2[i], ha='center')
  plt.text(x[i], y3[i]+0.2, y3[i], ha='center')

plt.title('기업별 매출액')
plt.legend(ncol=3) # 범례출력 기본: 세로출력, ncol = 3
plt.show()
```

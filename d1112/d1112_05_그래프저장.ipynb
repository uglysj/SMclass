```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Win용
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
# X축: 지원번호, Y축: 국어
# 선색상: red, marker색상: yellow, 테두리: blue, 선타입: --
```

```python
x = df.index
k = df['국어']
e = df['영어']
m = df['수학']

# 그래프 크기 조정, dpi: 그래프 해상도 설정, 해상도가 높을수록 이미지 크기 커짐, facecolor: 배경색설정
plt.figure(figsize=(20, 5), facecolor='yellow') # X축으로 5, Y축으로는 20크기조절
# plt.figure(figsize=(20, 5), dpi=300) # X축으로 5, Y축으로는 20크기조절

plt.title('과목')
plt.xlabel('지원번호')
plt.ylabel('점수')
plt.plot(x, k, 'ro--', mfc='yellow', mec='blue', label='국어')
plt.plot(x, e, 'gv--', mfc='#23bb00', mec='orange', label='영어')
plt.plot(x, m, 'bx--', mec='blue', label='수학')
plt.legend(loc='upper center')

# 그래프 파일로 저장
# plt.savefig('graph2.png', dpi=300)
```

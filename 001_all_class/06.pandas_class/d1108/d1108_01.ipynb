데이터 조건 검색

```python
# 조건식 생성
filt = df['키'] >= 185
filt
# 조건식을 대입
df[filt]
df[~filt] # not에 대한 결과값을 출력
df.loc[filt] # row
df.loc[filt, ['국어', '수학']] # row
df.loc[filt, '키'] # row
```

문자열 함수: startswith, contains, replace, lower, upper

```python
# startswith: 해당 값으로 시작되는 단어 검색
filt = df['이름'].str.startswith('김')
df[filt]

# contains: 해당 값이 포함되어 있는 단어 검색
filt = df['이름'].str.contains('수')
df.loc[filt, ['이름', '학교', '키']]

# isin() 포함이 되어 있는지 확인 / 괄호안에는 리스트 타입만 가능
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs)
df[filt]

# lower(): 해당값을 모두 소문자로 변환
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]

# 포함이 되어 있으면 모두 검색
# NaN 데이터 처리를 하지 않으면 에러
filt = df['SW특기'].str.contains('Java', na=False)
df[filt]
```

```python
filt = df['SW특기'].str.lower().isin(['java', 'python'])
df[filt]
```

```python
### j가 포함되어 있는 정보 검색
filt = df['SW특기'].str.lower().str.contains('j', na=False)
df[filt]
```

```python
filt = df['SW특기'].str.lower().isin(['python', 'java'])
df[filt]
```

```python
## dropna
# axis축방향으로 삭제: index(row방향), columns(컬럼방향)
# how: any (1개라도 있으면 모두 row나 columns삭제), all(모두 NaN데이터일 때 삭제)
# axis = 'index', 'columns', how='any', 'all'
df.dropna() # index, any 기본 default - 해당되는 index(row)를 삭제

df.dropna(axis='index', how='all') # 모두 NaN row가 없어서 삭제가 안됨.
df.dropna(axis='columns', how='all') # 학교 컬럼이 모두 NaN이기에 삭제됨.
df
df.dropna(axis='index', how='any') # 학교컬럼이 모두NaN 이기에 모든 row 삭제
df.dropna(axis='columns', how='any') # 학교, SW특기 컬럼에 NaN이 있기에 2개의 컬럼이 삭제. 
```

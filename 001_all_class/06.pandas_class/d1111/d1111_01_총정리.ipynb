```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
# 컬럼값 변경: 해당컬럼에 값을 입력하면 됨.
# 단 타입이 같아야 함.
df['학교'] = df['학교'] + '등학교'
df

# 같은 타입만 더하기 가능
# df['키'] + 'cm' # 에러
# df['키'].astype(str) + 'cm' # 가능
```

```python
# 구로고 학생만 국어점수 50으로 변경
filt = df['학교'] == '구로고'
# 컬럼으로 비교하고 값의 변경은 row 값을 찾아서 변경해야 함.
df.loc[filt, '국어'] = 50
```

```python
# 컬럼 추가
# 없는 컬럼명을 입력하면 컬럼이 추가 됨.
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
```

```python
c_list = list(df.columns)
```

```python
# type(c_list[0:8]) # list
# type(c_list[-1]) # str
# [1, 2] + [3, 4] # list 타입을 맞춰서 더하기를 해야 함.
```

```python
# 컬럼 순서 변경: 컬럼의 타입을 일치한 후 더하기를 해주면 됨.
# 리스트 + 리스트(문자열) + 리스트(문자열)
df = df[c_list[0:7] + [c_list[-1]] + [c_list[8]]]
df
```

```python
# 평균 컬럼을 넣어서, 이름 바로 뒤로 위치를 변경하시오.
# 없는 컬럼을 입력하면 추가, 있는 컬럼에 값을 입력하면 변경
df['평균'] = df['총합'] / 5
c_list = list(df.columns)
df = df[[c_list[0]] + [c_list[-1]] + c_list[1:-1]]
df
```

```python
# 결과 컬럼을 추가해서 평균 80점이상이면 합격, 불합격 출력
# 결과 컬럼 추가
# 필터 적용
# 합격, 불합격 입력
df['결과'] = '불합격'
filt = df['평균'] >= 80
df.loc[filt, '결과'] = '합격'
df
```

```python
## 컬럼 삭제: drop (columns), 적용: inplace=True
# df.drop(columns=['평균'], inplace=True)
df.drop(columns=['결과', '총합'], inplace=True)
df
```

```python
## index 삭제: drop (index), 적용: inplace=True
# df.drop(index=['1번']) # index를 가지고 삭제
df.drop(index=['1번', '3번']) # index를 가지고 삭제

# 수학점수가 80점이하 삭제
filt = df['수학'] < 80
# df.drop(index=df[filt].index)
df
```

```python
## 컬럼에 해당되는 값을 다른 값으로 변경: df.replace
## 컬럼값을 변경: df[컬럼]
## 컬럼추가: df[컬럼]
## 컬럼삭제: df.drop(columns), df.drop(index)
## 컬럼의 특정값을 변경: df.loc(조건, 컬럼) = 값
```

```python
# row 열을 추가
# 없는 index의 값을 가지고 입력 # df.index
df.loc['9번'] = ['홍길동', '디지털고', 191, 100, 89, 87, 67, 90, 'Django']
```

```python
### 50점 이하에 있는 학생의 점수를 +20
# filt = df['국어'] < 50
# df.loc[filt, '국어'] = df.loc[filt, '국어'] + 20
```

```python
### 함수
def func(value):
  if value < 50:
    return value + 20
  else:
    return value + 10

# 함수호출: apply()
df['국어'] = df['국어'].apply(func)
df
```

```python
# NaN 데이터 처리: fillna:입력, dropna: 삭제
# df.fillna('python') # inplace=True
```

```python
### 1. 소문자로 치환
### 2. NaN값에 python 데이터 입력
df['SW특기'] = df['SW특기'].str.lower()

# isnull: NaN 인지 확인, isnotnull(): NaN이 아닌지 확인
filt = df['SW특기'].isnull()
df.loc[filt, 'SW특기'] = 'python'
```

```python
### 함수
def func(value):
  if pd.notnull(value):
    return value.lower()
  else:
    return 'python'

df['SW특기'] = df['SW특기'].apply(func)
df
```

```python
import random
```

```python
###
# 1. 요일 컬럼 추가
df['요일'] = 0
# 2. 0 - 6 까지의 숫자를 랜덤으로 입력
for i in range(8):
  df.iloc[i, 9] = random.randint(0, 6)
# 3. 함수를 사용해서 0-일요일, 1-월요일, 2-화요일..... 6-토요일로 변경
def func(value):
  if value == 0: return '일요일'
  elif value == 1: return '월요일'
  elif value == 2: return '화요일'
  elif value == 3: return '수요일'
  elif value == 4: return '목요일'
  elif value == 5: return '금요일'
  else: return '토요일'

df['요일'] = df['요일'].apply(func)
```

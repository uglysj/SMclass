```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Win용
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

다중막대그래프

```python
# 배열 [0, 1, 2, 3, 4, 5, 6, 7]
index = np.arange(df.shape[0])
index
```

```python
w = 0.25
plt.bar(index, df['국어'], width=w, label='국어')
plt.bar(index-w, df['영어'], width=w, label='영어')
plt.bar(index+w, df['수학'], width=w, label='수학')
plt.xticks(index)
plt.ylim(0, 120)
plt.legend(ncol=3, loc='upper center')
for i in range(8):
  plt.text(index[i], df.iloc[i]['국어'], df.iloc[i]['국어'], ha='center')
  plt.text(index[i]-w, df.iloc[i]['영어'], df.iloc[i]['영어'], ha='center')
  plt.text(index[i]+w, df.iloc[i]['수학'], df.iloc[i]['수학'], ha='center')
plt.show()
```

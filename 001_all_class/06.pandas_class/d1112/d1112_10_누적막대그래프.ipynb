```python
import pandas as pd
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Win용
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

* 누적 막대그래프
* bottom: 첫번째 값을 입력

```python
plt.figure(figsize=(10, 4))
plt.bar(df['이름'], df['국어'], label='국어')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], label='영어')
plt.bar(df['이름'], df['수학'], bottom=df['국어'] + df['영어'], label='수학')
plt.xticks(rotation=45)
plt.legend(ncol=3, loc='upper center')
plt.ylim(0, 350)

for i in range(8):
  plt.text(df.iloc[i]['이름'], df.iloc[i]['국어'], df.iloc[i]['국어'], ha='center')
  plt.text(df.iloc[i]['이름'], df.iloc[i]['국어'] + df.iloc[i]['영어'], df.iloc[i]['영어'], ha='center')
  plt.text(df.iloc[i]['이름'], df.iloc[i]['국어'] + df.iloc[i]['영어'] + df.iloc[i]['수학'], df.iloc[i]['수학'], ha='center')

plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('chipotle.tsv', sep='\t')
df.tail(2)
```

```python
## 함수를 사용해서
## item_price $부분을 삭제하고, float 타입으로 변경해서 입력하시오.
def func(x):
  return float(x[1:])

df['item_price'] = df['item_price'].apply(func)
df['item_price'] = df['item_price'].apply(lambda x: float(x[1:]))
```

```python
df['item_price'].sum()
df['item_price'].mean()
df['item_price'].max()
df['item_price'].min()
```

```python
df.sort_values('item_price', ascending=False).head(30)
```

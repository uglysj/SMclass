```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df

# 학교 구로고 -> 가산고, 디지털고 -> 단지고 변경
# 4번 SW특기 -> Django
# 5번 SW특기 -> Python
# SW특기 소문자로 변경
# 합계컬럼을 추가하시오.
# 결과 컬럼을 추가해서 400점이상은 pass, 아니면 fail 출력

df['학교'].replace({'구로고': '가산고', '디지털고': '단지고'}, inplace=True)

df.loc[['4번', '5번'], 'SW특기'] = ['Django', 'Python']

df['SW특기'] = df['SW특기'].str.lower()

df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']

df['결과'] = 'Fail'
filt = df['합계'] >= 400
df.loc[filt, '결과'] = 'Pass'
```

```python
### 문자열 = 문자열 + 문자열
df['학교'] = df['학교'] + '등학교'
df
```
